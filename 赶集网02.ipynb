```python
def get_all_dates(date1,date2):
    '''
    year1='2018'
    year2='2018'
    if(int(date1.split('.')[0])>3):
        year1 = '2017'
    if(int(date1.split('.')[0])==3):
        if int(date1.split('.')[1])>17:
            year1 = '2017'
    
    if(int(date2.split('.')[0])>3):
        year2 = '2017'
    if(int(date2.split('.')[0])==3):
        if int(date2.split('.')[1])>17:
            year2 = '2017'
    '''
    
    the_date = date(2018,int(date1.split('.')[0]),int(date1.split('.')[1]))
    end_date = date(2018,int(date2.split('.')[0]),int(date1.split('.')[1]))
    days = timedelta(days=1)

    while the_date <= end_date:
        yield (the_date.strftime('%Y.%m.%d'))
        the_date = the_date + days
```

```python
for i in get_all_dates('03.04','12.17'):
    print(i)
```

```python
def get_data_within(date1,date2,areas):
    date1 = str(date1)[5:]
    date2 = str(date2)[5:]
    
    for area in areas:
        area_day_posts = []
        for date in get_all_dates(date1,date2):
            date = str(date)[5:]
            
            a = list(item_info.find({'pub_date':date,'area':area}))
            each_day_post = len(a)
            area_day_posts.append(each_day_post)
        data = {
            'name': area,
            'data': area_day_posts,
            'type': 'line'
        }
        yield data
```

```python
for i in get_data_within('2018.01.01','2018.03.17',['朝阳','海淀','通州']):
    print(i)
```

```python
options = {
    'chart'   : {'zoomType':'xy'},
    'title'   : {'text': '发帖量统计'},
    'subtitle': {'text': '可视化统计图表'},
    'xAxis'   : {'categories': [i for i in get_all_dates('01.01','03.17')]},
    'yAxis'   : {'title': {'text': '数量'}}
    }

series = [i for i in get_data_within('2018.01.01','2018.03.17',['朝阳','海淀','通州'])]


charts.plot(series, options=options,show='inline')
# options=dict(title=dict(text='Charts are AWESOME!!!'))
```

```python
%matplotlib inline
import pymongo
import charts
```

```python
client = pymongo.MongoClient('localhost', 27017)
ganji = client['ganji']
url_list = ganji['url_list01']
item_info = ganji['item_info01']
```

```python
for i in item_info.find({},{'_id':0,'cates1':1}).limit(300):
    print(i)
```

```python
pipeline = [
    {'$group':{'_id':'$price','counts':{'$sum':1}}},
    {'$sort' :{'counts':-1}},
    {'$limit':10}
]
```

```python
for i in item_info.aggregate(pipeline):
    print(i)
```

```python
def data_gen():
    pipeline = [
    {'$match':{'last_time':1}},
    {'$group':{'_id':'$cates1','counts':{'$sum':1}}},
    {'$sort':{'counts':-1}}
]
    for i in item_info.aggregate(pipeline):
        yield [i['_id'],i['counts']]
```

```python
for i in data_gen():
    print(i)
```

```python
options = {
    'chart'   : {'zoomType':'xy'},
    'title'   : {'text': '发帖量统计'},
    'subtitle': {'text': '二手物品类目分布占比'},
    }


series =  [{
    'type': 'pie',
    'name': 'pie charts',
    'data':[i for i in data_gen()]

        }]
charts.plot(series,options=options,show='inline')
```
